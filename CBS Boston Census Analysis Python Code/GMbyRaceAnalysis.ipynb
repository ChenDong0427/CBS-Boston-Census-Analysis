```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '/content/drive/MyDrive/Colab Notebooks/data'
```

```python
names = locals()
for i in range(10):
  names['df201' + str(i) ] = pd.read_csv(data_path+'/GMbyRace201{}.csv'.format(i))
```

```python
col_name = 'Massachusetts!!Moved; from different  state!!Estimate'
MA_move_state = []
for i in range(10):
  MA_move_state.append(names.get('df201'+str(i))[col_name])
```

```python
column_name = ('2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019')
plt.figure(figsize=(10, 8))

plt.plot(column_name, [float(row[1]) for row in MA_move_state], label='White')
plt.plot(column_name, [float(row[2].strip("%"))/100 for row in MA_move_state], label='Black or African American')
plt.plot(column_name, [float(row[4].strip("%"))/100 for row in MA_move_state], label='Asian')
plt.plot(column_name, [float(row[6].strip("%"))/100 for row in MA_move_state], label='Some other race')

plt.legend()
plt.title('Moving from different state for MA by Race')
plt.show()
plt.savefig("MA_state.png")
```
