## Import and extract data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# read dataset
bos_10 = pd.read_csv("./Tenure/Boston/ACSST1Y2010.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_11 = pd.read_csv("./Tenure/Boston/ACSST1Y2011.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_12 = pd.read_csv("./Tenure/Boston/ACSST1Y2012.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_13 = pd.read_csv("./Tenure/Boston/ACSST1Y2013.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_14 = pd.read_csv("./Tenure/Boston/ACSST1Y2014.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_15 = pd.read_csv("./Tenure/Boston/ACSST1Y2015.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_16 = pd.read_csv("./Tenure/Boston/ACSST1Y2016.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_17 = pd.read_csv("./Tenure/Boston/ACSST1Y2017.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_18 = pd.read_csv("./Tenure/Boston/ACSST1Y2018.S1101_data_with_overlays_2021-11-07T181626.csv")
bos_19 = pd.read_csv("./Tenure/Boston/ACSST1Y2019.S1101_data_with_overlays_2021-11-07T181626.csv")
```

```python
# bos contains all the datasets from 2010 to 2019
bos = [bos_10, bos_11, bos_12, bos_13, bos_14, bos_15, bos_16, bos_17, bos_18, bos_19]

owner_occupied_bos = [] #store owner occupied data
renter_occupied_bos = [] #store renter occupied data

# Extract corresponding data from the dataframe for year 2010 to 2018
for df in bos[:-1]:
    renter_occupied_bos.append(float(df['S1101_C01_020E'][1]))
    owner_occupied_bos.append(float(df['S1101_C01_019E'][1]))
    
# Extract corresponding data for year 2019, different table format from other years
renter_occupied_bos.append(float(bos[-1]['S1101_C01_018E'][1]))
owner_occupied_bos.append(float(bos[-1]['S1101_C01_017E'][1]))
```

## Boston

```python
#Plot for Boston housing tenure over the years
plt.style.use(['seaborn-paper', 'seaborn-whitegrid'])
plt.style.use(['seaborn'])
sns.set(palette='colorblind')

labels = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

bar_width = 0.20
data_bos = [owner_occupied_bos, renter_occupied_bos]

colors = sns.color_palette(palette='colorblind')
columns = ('2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019')

index = np.arange(len(labels))
fig = plt.figure(figsize=(12,9))
plt.bar(index, owner_occupied_bos, bar_width)
plt.bar(index+bar_width+.02, renter_occupied_bos, bar_width)
plt.table(cellText=data_bos,
          rowLabels=['owner', 'renter'],
          rowColours=colors,
          colLabels=columns,
          loc='bottom')


fig.subplots_adjust(bottom=0.05)
plt.xticks([])
plt.title('Housing Tenure % in Boston')
#plt.show()
plt.savefig("tenure_bos.png", dpi=200, bbox_inches = "tight")
```

```python
# Boxplot
labels = ['']
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

#Owner occupied
bp_with = ax1.boxplot(owner_occupied_bos,
                      showmeans=True,
                      vert=True,  
                      patch_artist=True,  
                      labels=labels)  
ax1.set_title('owner_occupied_bos')

#Renter occupied
bp_wo = ax2.boxplot(renter_occupied_bos,
                    showmeans=True,
                    vert=True,  
                    patch_artist=True, 
                    labels=labels)  
ax2.set_title('renter_occupied_bos')

#Set axis
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_ylabel('percentage')

#plt.show()
plt.savefig("bos_boxplot.png", dpi=200, bbox_inches = "tight")
```
